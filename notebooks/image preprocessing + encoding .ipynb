# Autoencoder

## Encoder

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
def build_encoder(latent_dimension):
    encoder=Sequential()
    
    encoder.add(Conv2D(8,(2,2), input_shape=(80,60,3), activation='relu'))
    encoder.add(MaxPooling2D(2))
    
    encoder.add(Conv2D(16, (2, 2), activation='relu'))
    encoder.add(MaxPooling2D(2))

    encoder.add(Conv2D(32, (2, 2), activation='relu'))
    encoder.add(MaxPooling2D(2))    
    
    encoder.add(Conv2D(64, (2, 2), activation='relu'))
    encoder.add(MaxPooling2D(2))
    
    encoder.add(Flatten())
    
    encoder.add(Dense(latent_dimension, activation='tanh'))
    
    return encoder
```

```python
encoder=build_encoder(10)
```

```python
designer_latent=encoder.predict(designer_data)
```

## Decoder 

```python
from tensorflow.keras.layers import Reshape, Conv2DTranspose
```

```python
(np.array([80, 60])/2)/2
```

```python
def build_decoder(latent_dimension):
    decoder=Sequential()
    
    decoder.add(Dense(20*15*10, activation='tanh', input_shape=(latent_dimension,)))
    
    decoder.add(Reshape((20,15,10)))

    decoder.add(Conv2DTranspose(3, (2, 2), strides=2, padding='same', activation='relu'))

    decoder.add(Conv2DTranspose(3, (2, 2), strides=2, padding='same', activation='relu'))
    
    return decoder
```

```python
decoder=build_decoder(10)
```

```python
decoder.summary()
```

```python
from tensorflow.keras.layers import Input
from tensorflow.keras import Model
from tensorflow.keras.callbacks import EarlyStopping
```

```python
def build_model(encoder, decoder):
    inp=Input((80,60,3))
    encoded=encoder(inp)
    decoded=decoder(encoded)
    autoencoder=Model(inp,decoded)
    
    autoencoder.compile(loss='mse',
              optimizer='adam')
    return autoencoder
```

```python
autoencoder=build_model(encoder, decoder)
```

```python
autoencoder.summary()
```

```python
autoencoder.fit(designer_data,
                designer_data, 
                epochs=20, 
                callbacks=EarlyStopping(patience=5, 
                                         restore_best_weights=True))
```
